## Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import PercentFormatter
sns.set()
```

```python
df_homework = pd.read_csv('bar_chart_homework_data.csv')
df_homework
```

## Horizontal Bar Chart

```python
plt.barh(y = df_homework['Cities'],
        width = df_homework['Frequency'],
        color = 'midnightblue'
       )
```

```python
plt.bar(x = df_homework['Cities'],
        height = df_homework['Frequency'],
        color = 'midnightblue'
       )
plt.xticks(rotation = 45 , fontsize = 11)
plt.title('Sales', color ='red', fontsize = 13, fontweight='bold')
plt.ylabel('Ammount', rotation = 0, loc='top')
```

## PIE CHART

```python
plt.pie(df_homework['Frequency'][0:3],
        labels = df_homework['Cities'][0:3]
       , textprops = {'color' : 'w'},
       autopct = '%.2f%%')
plt.legend()
plt.title('Pie Chart', fontsize = 13 , fontweight = 'bold')
sns.set_palette('colorblind')
plt.show()
```

## Area Chart Homework II - Create an Area Chart of a Company's Revenue

### Your homework task is to create a stacked area chart of a printer company and its revenue for the last 12 months. The data set contains monthly values in the span of a year and features three revenue sources:
### -	International Printer Sales
### -	Domestic Printer Sales
### -	Services (fixed fee)
### Think about the ordering of the categories, as well as the appropriate colors when displaying your chart. After you’ve completed the chart think about interpretation. What does the chart show? Which would you say is the most important revenue source for the business and why? How does the performance of the company look when considering the Date factor?



```python
area_df = pd.read_csv('area_chart_homework_data.csv', usecols=(0,1,2,3))
area_df.head(5)
```

```python
area_df.info()
```

```python
sns.set_style('white')
```

```python

labels = ['International', 'Domestic' , 'Service']
plt.figure(figsize = (15,4))

plt.stackplot(area_df['Date'],area_df['International'],area_df['Domestic'],area_df['Services'],
             edgecolor = 'none')
sns.despine()
plt.title('Printer Inc Revenue Report 2020', fontsize= 15)
plt.legend(labels = labels , loc='upper left')
plt.ylabel('Revenue in $')
plt.xticks(rotation = 45 , )
plt.show()
```

## Histogram Showing Distribution of Age

```python
df_hist = pd.read_csv('histogram_survey_data.csv')
df_hist
```

```python
sns.set_style('white')

plt.figure(figsize = (9,6))
plt.hist(df_hist['Age'],
        bins = 12 ,
        color = 'darkseagreen')
plt.title('Participants by Age', fontsize = 13 , fontweight= 'bold')
plt.xticks(np.arange(min(df_hist["Age"]), max(df_hist["Age"]),5.0))
plt.xlabel ('Age')
plt.ylabel ('Participants Number')
sns.despine()
plt.show()
```

## Bar and Line Chart Homework - Pareto Chart of Movie Goers' Complaints

```python
df_pareto = pd.read_csv('bar_line_chart_homework.csv')
df_pareto
```

```python
sns.set_style('white')

fig , ax = plt.subplots( figsize = (12,6))
ax.bar (df_pareto['Movie Theater Goers Complaints'] , df_pareto['Number of complaints'],
       color = 'blue')
ax.set_ylabel ( 'Number of Complaints', fontweight = 'bold')
ax.tick_params ( axis = 'y' , width = 2, )
ax1 = ax.twinx()
ax1.set_ylim(0,1.1)
ax1.yaxis.set_major_formatter(
PercentFormatter(xmax= 1))
ax1.plot( df_pareto['Movie Theater Goers Complaints'], df_pareto['frequency'],
        color = '#b60000',
        marker = 'D')
ax1.set_ylabel ( 'Frequency', fontweight = 'bold',  color = '#b60000')
ax1.tick_params ( axis = 'y' , width = 2, )
ax1.set_title ('Top 5 Reason of Movie Theater Complaints', fontweight = 'bold' , fontsize =15)
sns.despine(top = True, right =False)
```
